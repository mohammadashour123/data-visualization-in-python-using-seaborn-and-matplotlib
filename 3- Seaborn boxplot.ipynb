# Seaborn: boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
```

```python
cars = sns.load_dataset('mpg')

cars.shape
```

```python
cars = cars.dropna()
```

```python
cars.shape
```

```python
cars.head()
```

## Basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
cars.shape
```

```python
sns.boxplot(cars.mpg);
```

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.origin, y=cars.mpg);
```

```python
sns.boxplot(x='origin', y='mpg', data=cars);
```

### Hue

#### By Cylinders

```python
sns.boxplot(x='origin', y='mpg', hue='cylinders', data=cars);
```

## Styling Options

### order

```python
sns.boxplot(x='mpg', y='origin',   
            data=cars
           );
```

```python
sns.boxplot(x='mpg', y='origin', 
            data=cars,
            order=['japan', 'europe', 'usa']
           );
```

### color

```python
sns.boxplot(x='mpg', y='origin',  
            data=cars, 
            color='r'
           );
```

```python
sns.boxplot(x='mpg', y='origin', 
            data=cars,
            color='g'
           );
```

### width

```python
sns.boxplot(x='mpg', y='origin', 
            data=cars,
            width=0.6
           ); 
```

### linewidth

```python
sns.boxplot(x='mpg', y='origin', 
            data=cars,
            linewidth=5.5
           );
```

### whis

```python
sns.boxplot(x='mpg', y='origin', 
            data=cars,
            whis=1
           );
```

```python
sns.boxplot(x='mpg', y='origin', 
            data=cars,
            whis=5 
           );
```

### fliersize

```python
sns.boxplot(x='mpg', y='origin', 
            data=cars,
            fliersize=2
           );
```

```python
sns.boxplot(x='mpg', y='origin', 
            data=cars,
            fliersize=20
           );
```
