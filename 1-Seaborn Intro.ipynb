## seaborn vs matplotlib

* Seaborn allows you to visualize regression models while matplotlib does not.
* Matplotlib has a more extensive library of functions than seaborn. 
* However, seaborn is growing at a faster rate than matplotlib.


# Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
x = [10, 20, 30, 40]
y = [0, 15, 10, 25]
```

```python
plt.plot(x, y);
```

# Seaborn

Seaborn is a library mostly used for statistical plotting in Python. It is built on top of Matplotlib and provides beautiful default styles and color palettes to make statistical plots more attractive.
