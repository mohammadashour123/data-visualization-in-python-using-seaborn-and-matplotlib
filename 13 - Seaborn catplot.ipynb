# Seaborn : catplot

```python
import seaborn as sns
```

```python
# load car data and drop rows with missings
cars = sns.load_dataset('mpg').dropna()

# filter to 4-, 6-, 8-cylinder cars
cars = cars[cars.cylinders.isin([4, 6, 8])] 

# add old/new column
cars['type'] = ['old' if x <=76 else 'new' for x in cars.model_year]
```

```python
cars.head()
```

```python
cars.shape
```

## Basics

```python
sns.set_style('white')
```

```python
sns.catplot(x='horsepower', data=cars);
```

```python
sns.catplot(y='horsepower', data=cars);
```

### Kind

```python
sns.catplot(x='horsepower', data=cars, kind='box');
```

```python
sns.catplot(x='horsepower', data=cars, kind='violin');
```

### Categorical Variables, Hue

```python
sns.catplot(x='origin', y='horsepower', 
            data=cars, 
            kind='box');
```

```python
sns.catplot(x='origin', y='horsepower', 
            data=cars, 
            hue='cylinders',
            kind='box');
```

```python
sns.catplot(x='origin', y='horsepower',
            data=cars, 
            hue='cylinders',
            kind='strip'); 
```

## Small Multiples

```python
sns.set_style('darkgrid')
```

#### `row`, `col`

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box',
            col='origin'
           );
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin'
           );
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            row='type'
           );
```

#### `row_order` (`order`, `hue_order`, `col_order`)

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            row='type',
            row_order=['new', 'old']
           );
```

## Styling

In general, additional keyword arguments will be passed to the kind of plot selected.  

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box',
            col='origin',
            palette='winter', 
            whis=1
           );
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            s=12,
            alpha=0.8,
            linewidth=8
           );
```
