# Seaborn: stripplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

## Basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
cars.cylinders.value_counts()
```

```python
sns.stripplot(cars.weight);
```

```python
sns.stripplot(cars.weight, cars.origin);
```

```python
sns.stripplot(x ='weight', y='origin', hue='cylinders',
              data=cars
             );
```

```python
sns.stripplot(x ='weight', y='origin', hue='cylinders', 
              data=cars,
              dodge=True
             );
```

### Orientation

```python
sns.stripplot(x='origin', y='weight', data=cars);
```

## Styling

```python
sns.set_style('white')
```

### jitter

```python
sns.stripplot(cars.weight, cars.origin, jitter=False);
```

### alpha

```python
sns.set_style('whitegrid')
```

```python
sns.stripplot(cars.weight, cars.origin, alpha=0.3);
```

### size

```python
sns.stripplot(cars.weight, cars.origin, 
              alpha=1,
              size=8    #Default size is 5
             );
```

### linewidth, edgecolor

```python
sns.stripplot(cars.weight, cars.origin, linewidth=5);
```

```python
sns.stripplot(cars.weight, cars.origin, linewidth=3, edgecolor='black');
```

### marker



```python
sns.stripplot(cars.weight, cars.origin, 
              alpha=0.7, size=10, jitter=0.25, marker='*'
             );
```
