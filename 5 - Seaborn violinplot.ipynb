# Seaborn: violinplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

## Basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(cars.displacement);
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement);
```

```python
sns.violinplot(x='cylinders', y='displacement',
               data=cars
              );
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars
              );
```

## Violinplot Options

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])]
              )

plt.legend(loc=2);
```

### split

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True
              )

plt.legend(loc=5);
```

### inner

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               inner='quartiles'
              )

plt.legend(loc=2);
```

### scale

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               scale='count'
              )

plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               scale='count',
               scale_hue=False
              )

plt.legend(loc=8);
```

```python
(
    cars[cars.origin.isin(['japan', 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```
